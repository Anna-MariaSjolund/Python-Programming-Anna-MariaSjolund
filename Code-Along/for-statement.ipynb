# for-loop

- "foreach" loop
- for i in range(10) in python (range 10 means 0-9)
- in some other languages: for (int i = 0; i<10; i++)

```python
for i in range(10):
    print(i, end = " ")

print() #new line

#every third number
for i in range(1,20,3):
    print(i, end = " ")

print(f"\n{range(10)}") #range is not a list
print(f"En lista från range(10): {list(range(10))}") #we can create a list from range
```

```python
#Sum all numbers between 0-5

sum_ = 0 #There exists a in built function sum() in Python. Therefore we will name it sum_

for i in range(6): #Will print 0+1+2+3+4
    sum_ += i
    if i == 5:
        continue #Skips the current iteration
    print(f"{i}+", end="")

print(f"{i}={sum_}") #Prints 5=15 (adds it directly after 0+1+2+3+4)
print(sum(range(6))) #Uses python's sum() function with range()
```

## Dice simulations
- calculate probability of dice roll 6
- P(roll 6) = number of sixes/total number of rolls

```python
import random as rnd

num_six, dice_rolls = 0, 1000000

#Count number of six
for i in range(dice_rolls):
    dice = rnd.randint(1, 6)
    if dice == 6:
        num_six += 1

print(f"Number of six is {num_six} in {dice_rolls} dice rolls.")
print(f"Probability of six is {num_six/dice_rolls}.")
print(f"Theoretical probability is {1/6}.")
```

## Loop through list

```python
fruits = {"apple", "orange", "strawberry", "grape"} #created a list of yummy fruits
print("I love to eat: ")

for fruit in fruits:
    print(fruit)
```

```python
fruits = {"apple", "orange", "strawberry", "grape"} #created a list of yummy fruits
print("Shopping list: ")

#When we use enumerate we will also get the index
for i, fruit in enumerate(fruits): #is the index, fruit is the element
    print(f"{i}. {fruit}")
```

## Nationella prov - Ma3C pastauppgift

```python
#Ma3C HT14 uppg 18

import math

for year in range(400): #We do not know which year it should stop (should probably use a while-statement instead)
    pasta_consumption = 0.791*math.exp(0.0526*year)
    if pasta_consumption >= 15:
        print(f"År {year + 1960} är pastakonsumptionen 15 kg.")
        break
```

```python
import matplotlib.pyplot as plt

pasta_per_year = []

for year in range(140):
    pasta_consumption = 0.791*math.exp(0.0526*year)
    pasta_per_year.append(pasta_consumption)

plt.figure(dpi = 100)
plt.plot(range(1960, 2100), pasta_per_year)
plt.xlabel("År")
plt.ylabel("Pastakonsumption")
plt.title("Pastakonsumptionsmodell per kg per person från 1960")
plt.show() #Måste vara med om man kör python-skript

print("Orimligt mycket pasta!")
```

## Nested loops
- A loop in a loop

```python
for row in range(0,3):
    for col in range(0,3):
        print(f"({row}, {col})", end = "")
    print()
```
