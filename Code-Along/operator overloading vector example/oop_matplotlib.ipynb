# Matplotlib
- MATLAB-style
- OOP appoach

```python
# Setup
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-5, 5) #Equally spaced values between -5 and 5 (we can also specify how many values we want)
f = np.sin(x)

#print(x)
```

## MATLAB style interface

```python
plt.figure(dpi = 100) 
plt.plot(x, f) #Sinus curve
plt.title("sin(x)")
plt.xlabel("x")
plt.ylabel("y")
```

```python
plt.figure(dpi = 100, figsize=(10, 3))
plt.subplot(1, 2, 1) #(rows, cols, panel number)
plt.plot(x, np.sin(x))
plt.plot(x, np.cos(x)) #Add another plot in the same window

plt.subplot(1, 2, 2)
plt.plot(x, np.cos(x))
#plt.show() stops the program, in a python script
```

## OOP approach

```python
#There is a figure class in matplotlib, even though matplotlib is not a class(?)
#We can create an object
fig = plt.figure(dpi=100) #command click to show documentation for figure 
ax = plt.axes() #ax object can have a plot, axis, labels, title, ...
print(fig) #A figure object (600x400), size is dependent on dpi

ax.plot(x, f) #We can now use ax.plot() instead of plt.plot()
#$ LaTeX (spelling?) code, for writing mathematical functions
ax.set(xlabel="x", ylabel="y", title="$\sin{x}$", xlim=(0,5)); #We can disable output by adding ;
```

```python
fig, ax = plt.subplots(1, 2, dpi=100, figsize=(10, 4)) #Creates two empty plots #subplots returns a tuple
rng = np.random.RandomState(42) #we set a seed and generates random numbers
data_normal_dist = rng.randn(1000) #Generates 1000 random number, normally distributed
ax[0].hist(data_normal_dist, bins=30, alpha=.5); #ax[0] #this acts on the first plot #alpha is transperacy
ax[0].set(title="Histogram of normal distributed data", xlabel="x", ylabel="y")

bubble = ax[1].scatter(x=rng.randn(100), y=rng.randn(100), c=rng.rand(100), alpha=.7, cmap="cividis", s=100*rng.rand(100)) #We randomise the colours
ax[1].set(title="Bubble chart", xlabel="x", ylabel="y")
fig.colorbar(bubble)

plt.savefig("beautiful_plots.png", facecolor="white") #Saves the plot in png-format
```
