## 1. Dice rolls

```python
import random as rnd

#Create a list of random dice rolls and print it out
dice_rolls = [rnd.randint(1, 6) for _ in range(10)]
print(dice_rolls)

#Sort the list in ascending order
dice_rolls.sort()
print(f"Ascending order: {dice_rolls}.")

#Sort the list in descending order
dice_rolls.sort(reverse=True)
print(f"Descending order: {dice_rolls}.")

#find the maximum and minimun value of the list
print(f"The maximum value for the dice rolls were: {max(dice_rolls)}.")
print(f"The minimum value for the dice rolls were: {min(dice_rolls)}.")
```

## 3. Squares

```python
import matplotlib.pyplot as plt

#Create a list of the squared numbers -10 to -10 
squares = [number**2 for number in range(-10, 11)]
print(squares)

#Create a plot of the function x^2
plt.figure(dpi = 100)
plt.title("The function x^2")
plt.plot(range(-10, 11), squares)
plt.xlabel("Numbers")
plt.ylabel("Squared Numbers")
```

## 4. Chessboard

```python
#Creates a list of upper case letters together with 1:s
#chr generates a character (each character has a number.
#i takes values between 97 and 105. 97 means "a" and 105 means "h"
#.upper converts the letters to uppercase.
name_of_squares = [f"{chr(i).upper()}1" for i in range(97, 105)] 
print(name_of_squares)

#New line
print()

#Creates a chessboard matrix (see code above)
#for j in range is the eight rows (j vill be used for the row's number)
chessboard = [[f"{chr(letter).upper()}{number}" for letter in range(97, 105)] for number in range(8, 0, -1)]

# * symbol is used to print the list elements in a single line with space. 
#To print all elements in new lines use sep=”\n”
print(*chessboard, sep = "\n")
```

## 5. Dice rolls convergence

```python
#Import libraries
import random as rnd
import matplotlib.pyplot as plt
import matplotlib.ticker as plticker
from math import log

#Create 100 random integers between 1 and 6 and count the number of sixes
no_of_sixes = [rnd.randint(1, 6) for _ in range(100)].count(6)


#Creates a list with the number of rolls
number_of_rolls = [10, 100, 1000, 10000, 100000, 1000000]

#Creates random integers between 1 and 6 and count the number of sixes 
#It does this for all of the different number of rolls
number_of_sixes = [[rnd.randint(1, 6) for _ in range(rolls)].count(6) for rolls in number_of_rolls]
#print(number_of_sixes)

#Counts the probability (number of sixes/number of rolls) for the different number of rolls
probability = [(number_of_sixes[i]/number_of_rolls[i]) for i in range(len(number_of_rolls))]
#print(probability)

#Create a plot, showing number of sixes and probability
fig = plt.figure(dpi = 100)
plt.title("Probability of Getting a Six When Throwing Dice")

log_rolls = list(map(log,number_of_rolls)) #Map takes a function (i.e. log) and applies it on all the values in number_of_rolls and then saves them to a list.

a = plt.plot(log_rolls, probability) #Plots the log values on the x-axis and probabibility on the y-axis
b = plt.xticks(log_rolls, number_of_rolls) #The first argument takes the number of ticks and the second argument overwrites the old tick labels

#This code can also be used instead of: log_rolls = list(map(log,number_of_rolls))
#my_logs = []
#logger = lambda x : my_logs.append(log(x))

#for x in number_of_rolls:
    #logger(x)
#print(my_logs)


```
