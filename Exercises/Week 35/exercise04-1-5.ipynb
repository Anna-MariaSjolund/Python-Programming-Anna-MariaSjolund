## 1. Dice rolls

```python
import random as rnd

#Create a list of random dice rolls and print it out
dice_rolls = [rnd.randint(1, 6) for _ in range(10)]
print(dice_rolls)

#Sort the list in ascending order
dice_rolls.sort()
print(f"Ascending order: {dice_rolls}.")

#Sort the list in descending order
dice_rolls.sort(reverse=True)
print(f"Descending order: {dice_rolls}.")

#find the maximum and minimun value of the list
print(f"The maximum value for the dice rolls were: {max(dice_rolls)}.")
print(f"The minimum value for the dice rolls were: {min(dice_rolls)}.")
```

## 2. Food menu

```python
#Creates two lists: menu and weekdays
dishes = ["Vegetarisk lasagne", "Spaghetti", "Fisk", "Grönsakssoppa", "Pannkakor"]
weekdays = ["Måndag", "Tisdag", "Onsdag", "Torsdag", "Fredag"]

#Creates one list of lists (the lists contain weekday and the dish of the day)
weekly_menu = ["Måndag", "Vegetarisk lasagne"], ["Tisdag", "Spaghetti"], ["Onsdag", "Fisk"], ["Torsdag", "Grönsakssoppa"], ["Fredag", "Pannkakor"]

#Print a menu, using the list weekly_menu
print("Veckomeny:")
for i in range(5):
    print(f"{weekly_menu[i][0]}: {weekly_menu[i][1]}")

print()

#Print a menu, using the lists menu and weekdays
print("Veckomeny")
for i in range(5):
    print(f"{weekdays[i]}: {dishes[i]}")
```

## 3. Squares

```python
import matplotlib.pyplot as plt

#Create a list of the squared numbers -10 to -10 
squares = [number**2 for number in range(-10, 11)]
print(squares)

#Create a plot of the function x^2
plt.figure(dpi = 100)
plt.title("The function x^2")
plt.plot(range(-10, 11), squares)
plt.xlabel("Numbers")
plt.ylabel("Squared Numbers")
```

```python
import random as rnd

def number_of_sixes(number_of_rolls):
    for i in range(len(number_of_rolls)):
        rolls = [rnd.randint(1, 6) for i in range(number_of_rolls[i])]
        print(rolls.count(6))

list_ = [1, 100, 1000, 100000, 100000]
number_of_sixes(list_)


```
